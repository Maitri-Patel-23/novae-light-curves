```python
# https://projectpluto.com/galnovae/galnovae.htm koji's list 
# https://www.aavso.org/vsx/index.php?view=detail.top&oid=8278029 V7992 Sgr  link

#2025
# V7992 Sgr 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load space-delimited AAVSO data
file_path = r"C:\Users\Novea research\1\V7992 Sgr.txt"

df = pd.read_csv(file_path, delim_whitespace=True)

# Extract Julian Date (JD) and Magnitude
jd = df["JD"]
mag = df["Magnitude"]

# Plot the light curve
plt.figure(figsize=(5, 2))
plt.scatter(jd, mag, c='red', marker='o', label="Nova Light Curve")
plt.gca().invert_yaxis()  # Brighter objects have lower magnitudes
plt.xlabel("Julian Date (JD)")
plt.ylabel("Magnitude")
plt.title("Light Curve of the Novae V7992 Sgr")
plt.legend()
plt.grid()
plt.show()

# Find peak brightness (lowest magnitude)
min_mag = mag.min()
jd_peak = jd[mag.idxmin()]

# Compute t2 and t3 magnitude levels
t2_mag = min_mag + 2
t3_mag = min_mag + 3

# Find the first Julian Date where the nova dims to t2 and t3 magnitudes
jd_t2 = jd[mag >= t2_mag].min()
jd_t3 = jd[mag >= t3_mag].min()

# Compute t2 and t3 values
t2 = jd_t2 - jd_peak if not np.isnan(jd_t2) else None
t3 = jd_t3 - jd_peak if not np.isnan(jd_t3) else None

t2, t3
```
